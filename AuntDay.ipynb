```python
names = ['Sally','Ava','Millie','Roxanne','Anna','Maryam','Paige']
graph = {
  'Sally' : ["Ava","Maryam","Paige"],
  'Ava' : ["Sally","Roxanne"],
  'Millie' : ["Paige"],
  'Roxanne' : ["Ava","Anna"],
  'Anna' : ["Roxanne","Maryam"],
  'Maryam' : ["Anna","Sally"],
  'Paige' : ["Sally","Millie"]
}

# ['Sally', 'Roxanne', 'Millie']
# [ Ava, Paige , Anna]
# both of group does not like maryam

visited = set()

table1 = []
table2 = []

def dfs(visited, graph, node):  #function for dfs 
    flag = 0  
    flag2 = 0 
    if node not in visited:
        print (node)
        visited.add(node)
        if node not in table1:
            for i in graph[node]:
                if i in table1:
                    flag = 0
                    break
                flag = 1
        if flag == 1 and node not in table1:
            table1.append(node)
        if node not in table2 and node not in table1:
            for i in graph[node]:
                if i in table2:
                    flag2 = 0
                    break
                flag2 = 1
        if flag2 == 1 and node not in table2:
            table2.append(node)
        flag = 0
        flag2 = 0
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

# Driver Code
print("= Depth-First Search = ")
dfs(visited, graph, 'Sally')

# make it list and retrieve data
# Sally dislikes = Ava Maryam Paige
# Ava dislikes =  Sally Roxanne
# Millie = Paige
# Roxanne = Ava Anna
# Anna = Roxanne Maryam
# Maryam Anna Sally
# Paige Sally Millie

bothofDoesNotLike = []

for i in names:
    if i not in table1 and i not in table2:
        bothofDoesNotLike.append(i)
    
print("Table 1 = ",table1)
print("Table 2 = ",table2)
print("Both of them does not like him = ",bothofDoesNotLike)

```
